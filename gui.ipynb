```python

from tkinter import *
import random

janela = Tk()

janela.geometry("175x250+550+360")
janela.config(background = "black")

lista_locais = []
lista_produto_posicao = [2,2,2,2,2,2,2,2,2]

a1 = Button(text = " ", height = 2, width = 4, background = "dark orange", command = lambda : setar(a1))
a1.place(x = 10, y = 60)

a2 = Button(text = " ", height = 2, width = 4, background = "blue", command = lambda : setar(a2))
a2.place(x = 60, y = 60)

a3 = Button(text = " ", height = 2, width = 4, background = "green", command = lambda : setar(a3))
a3.place(x = 110, y = 60)

b1 = Button(text = " ", height = 2, width = 4, background = "red", command = lambda : setar(b1))
b1.place(x = 10, y = 110)

b2 = Button(text = " ", height = 2, width = 4, background = "dark gray", command = lambda : setar(b2))
b2.place(x = 60, y = 110)

b3 = Button(text = " ", height = 2, width = 4, background = "yellow", command = lambda : setar(b3))
b3.place(x = 110, y = 110)

c1 = Button(text = " ", height = 2, width = 4, background = "purple", command = lambda : setar(c1))
c1.place(x = 10, y = 160)

c2 = Button(text = " ", height = 2, width = 4, background = "pink", command = lambda : setar(c2))
c2.place(x = 60, y = 160)

c3 = Button(text = " ", height = 2, width = 4, background = "dark red", command = lambda : setar(c3))
c3.place(x = 110, y = 160)


reset = Button(text = "Reset", height = 2, command = lambda: reset_game(lista_locais, lista_produto_posicao)
reset.place(x = 60, y = 10)


lista_button = [a1,a2,a3,b1,b2,b3,c1,c2,c3]

def reset_game(lista_1, lista_2):
    lista_1 = []
    lista_2 = [2,2,2,2,2,2,2,2,2]
    for i in lista_button:
        i.configure(text = " ", state = ACTIVE)



### Function ##

def setar(button):
    button.configure(text = "X", state = DISABLED)
    lista_locais.append(button)
    print(lista_locais)
    print(len(lista_locais))

    first_round(button)

def first_round(button):
    if len(lista_locais) ==1:
        if len(lista_locais) == 1 and lista_locais.count(b2) == 1:
            after_random = random.choice([a1, a3, c1, c3])
            after_random.configure(text = "O", state=DISABLED)
            lista_locais.append(button)
        elif len(lista_locais) == 1 and lista_locais.count(b2) == 0:
            button = b2
            button.configure(text = "O", state=DISABLED)
            lista_locais.append(button)
    else:
        analise_game()

def analise_game():
    for i in range(len(lista_button)):
        if lista_button[i].cget("text") == "X":
            lista_produto_posicao[i] = 3
        elif lista_button[i].cget("text") == "O":
            lista_produto_posicao[i] = 4
        else:
            lista_produto_posicao[i] = 2
    cabe_vitoria = attack()
    cabe_derrota = defesa()
    jogada_i_a(cabe_vitoria, cabe_derrota, lista_produto_posicao)

def attack():
    victory_case = "sem_vitoria"
    if lista_produto_posicao[0] * lista_produto_posicao [1] * lista_produto_posicao [2] ==32:
        victory_case = "case1"
        return victory_case
    elif lista_produto_posicao[3] * lista_produto_posicao [4] * lista_produto_posicao [5] ==32:
        victory_case = "case2"
        return victory_case
    elif lista_produto_posicao[6] * lista_produto_posicao [7] * lista_produto_posicao [8] ==32:
        victory_case = "case3"
        return victory_case
    elif lista_produto_posicao[0] * lista_produto_posicao [3] * lista_produto_posicao [6] ==32:
        victory_case = "case4"
        return victory_case 
    elif lista_produto_posicao[1] * lista_produto_posicao [4] * lista_produto_posicao [7] ==32:
        victory_case = "case5"
        return victory_case
    elif lista_produto_posicao[2] * lista_produto_posicao [5] * lista_produto_posicao [8] ==32:
        victory_case = "case6"
        return victory_case 
    elif lista_produto_posicao[0] * lista_produto_posicao [4] * lista_produto_posicao [8] ==32:
        victory_case = "case7"
        return victory_case
    elif lista_produto_posicao[2] * lista_produto_posicao [4] * lista_produto_posicao [6] ==32:
        victory_case = "case8"
        return victory_case
    else:
        victory_case = "sem_vitoria"
        return victory_case



def defesa():
    derrota_case = "sem_derrota"
    if lista_produto_posicao[0] * lista_produto_posicao [1] * lista_produto_posicao [2] ==18:
        derrota_case = "case1"
        return derrota_case
    elif lista_produto_posicao[3] * lista_produto_posicao [4] * lista_produto_posicao [5] ==18:
        derrota_case = "case2"
        return derrota_case
    elif lista_produto_posicao[6] * lista_produto_posicao [7] * lista_produto_posicao [8] ==18:
        derrota_case = "case3"
        return derrota_case
    elif lista_produto_posicao[0] * lista_produto_posicao [3] * lista_produto_posicao [6] ==18:
        derrota_case = "case4"
        return derrota_case 
    elif lista_produto_posicao[1] * lista_produto_posicao [4] * lista_produto_posicao [7] ==18:
        derrota_case = "case5"
        return derrota_case
    elif lista_produto_posicao[2] * lista_produto_posicao [5] * lista_produto_posicao [8] ==18:
        derrota_case = "case6"
        return derrota_case 
    elif lista_produto_posicao[0] * lista_produto_posicao [4] * lista_produto_posicao [8] ==18:
        derrota_case = "case7"
        return derrota_case
    elif lista_produto_posicao[2] * lista_produto_posicao [4] * lista_produto_posicao [6] ==18:
        derrota_case = "case8"
        return derrota_case
    else:
        derrota_case = "sem_derrota"
        return derrota_case


def jogada_i_a(win, loss, vet_vazio):
    print(win)
    print(loss)
    if win == "sem_vitoria" and loss == "sem_derrota":
        vet_escolha = []
        for i in vet_vazio:
            if lista_produto_posicao[i] == 2:
                vet_escolha.append(i)
        if len(vet_escolha) == 0:
            for i in range(len(lista_produto_posicao)):
                if lista_produto_posicao[i] == 2:
                    vet_escolha.append(i)
            escolha = random.choice(vet_escolha)
            lista_button[escolha].configure(text = "O", state = DISABLED)
        else:
            escolha = random.choise(vet_escolha)
            lista_button[escolha]. configure (text = "O", state = DISABLED)

    #Ganhando
    elif win != "sem_vitoria":
        if win == "case1":
            vet = [0,1,2]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure(text = "O", state = DISABLED)
        elif win == "case2":
            vet = [3,4,5]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif win == "case3":
            vet = [6,7,8]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif win == "case4":
            vet = [0,3,6]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif win == "case5":
            vet = [1,4,7]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif win == "case6":
            vet = [2,5,8]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif win == "case7":
            vet = [0,4,8]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif win == "case8":
            vet = [2,4,6]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)

    #Defendendo
    elif loss != "sem_derrota":
        if loss == "case1":
            vet = [0,1,2]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif loss == "case2":
            vet = [3,4,5]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif loss == "case3":
            vet = [6,7,8]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif loss == "case4":
            vet = [0,3,6]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif loss == "case5":
            vet = [1,4,7]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif loss == "case6":
            vet = [2,5,8]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif loss == "case7":
            vet = [0,4,8]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)
        elif loss == "case8":
            vet = [2,4,6]
            for i in vet:
                if lista_produto_posicao[i] == 2:
                    lista_button[i].configure (text = "O", state = DISABLED)

janela.mainloop()
```
